```python
import pandas as pd
from pandas.io.html import read_html
from selenium import webdriver
from datetime import datetime
```

```python
pages = ['https://www.f1fantasytracker.com/sentiment-tracker.html', 'https://www.f1fantasytracker.com/prices.html']
ids = ['constructorTable', 'driverTable']
```

```python
from webdriver_manager.firefox import GeckoDriverManager

driver = webdriver.Firefox(executable_path=GeckoDriverManager().install())
# driver.get()
```

```python
frames = []

for page in pages:
    driver.get(page)
    for id in ids:
        table = driver.find_element_by_id(id)
        data = read_html(table.get_attribute('outerHTML'))
        frames.append(data[0])
        
```

```python
driver.close()
driver.quit()
```

```python
driver = frames[1].merge(frames[3], on='Driver')
```

```python
team = frames[0].merge(frames[2], on='Team')
```

```python
out = [team, driver]

for frame in out:
    frame['ts'] = (datetime.now()).strftime('%Y-%m-%d %H:%M:%S')
```

## Write Data

```python
frame_labels = ['team_sent', 'driver_sent']

for i, frame in enumerate(out):
    frame.to_csv(f'./data/raw/sent/{frame_labels[i]}-{datetime.now():%Y-%m-%d.%H%M%S}.csv', sep='|')
    
```
