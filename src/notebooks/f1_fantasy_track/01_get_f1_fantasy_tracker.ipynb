```python
import pandas as pd
from pandas.io.html import read_html
from selenium import webdriver
import time
```

```python
pages = ['https://www.f1fantasytracker.com/standings.html', 'https://www.f1fantasytracker.com/prices.html']
ids = ['constructorTable', 'driverTable']
```

```python
from webdriver_manager.firefox import GeckoDriverManager

driver = webdriver.Firefox(executable_path=GeckoDriverManager().install())
# driver.get()
```

```python
frames = []
tables = []

for page in pages:
    driver.get(page)
    time.sleep(1)
    
    for id in ids:
        table = driver.find_element_by_id(id)
        tables.append(table)
        data = read_html(table.get_attribute('outerHTML'))
        frames.append(data[0])
        
```

```python
driver.close()
driver.quit()
```

```python
frame_labels = ['team_pts', 'driver_pts', 'team_price', 'driver_price']

for i, frame in enumerate(frames):
    frame.to_csv(f'./data/raw/{frame_labels[i]}.csv', sep='|')
    
```
